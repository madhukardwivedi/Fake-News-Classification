```python
df = pd.read_csv('/content/drive/MyDrive/train.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
## Dependent and Independent features

X= df.drop('label', axis=1)
y = df['label']
```

```python
X.shape
```

```python
y.shape
```

```python
# copy the data
news = df.copy()
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news['text'][2]
```

```python
## Now we will clean the data
import re
nltk.download('stopwords')
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords

ps = PorterStemmer()

corpus = []

for i in range(0, len(news)):
    review = re.sub('[^a-zA-Z]', ' ', news['text'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus[0]
```

```python
## Now create the count vectorizer using TF-IDF

from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=5000, ngram_range=(1,3))
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
y = news['label']
```

```python
## Divide the data into train and test 

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
cv.get_feature_names()
```

```python
cv.get_params()
```

```python
count_df = pd.DataFrame(X_train, columns=cv.get_feature_names())
count_df.head()
```

## Machine Learning Model

### Naive Bayes

```python
## Now our dataset is ready. We can apply the algorithms

## Naive Bayes

from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB()
```

```python
model = classifier.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test, pred)
cm = confusion_matrix(y_test, pred)
```

```python
print('accuracy is:\n ', accuracy)
print('\n confusion matrix is: ', cm)
```

```python
feature_names = cv.get_feature_names()
```

```python
## Most real words
sorted(zip(model.coef_[0], feature_names), reverse=True)[:20]
```

```python
## Most fake
sorted(zip(model.coef_[0], feature_names))[:5000]
```

### Random forest classifier

```python
### Random forest

from sklearn.ensemble import RandomForestClassifier
Rf = RandomForestClassifier(n_estimators=200, criterion='entropy')
```

```python
rf_model = Rf.fit(X_train, y_train)
```

```python
rf_pred = rf_model.predict(X_test)
```

```python
accuracy_rf = accuracy_score(y_test, rf_pred)
cm_rf = confusion_matrix(y_test, rf_pred)
```
